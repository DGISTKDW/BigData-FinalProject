## Blood Press EDA with smoke and age

```python
#2009
bp_2009 = GJ_2009[['smoke','age','b_press_contract','b_press_relax']]
bp_2009_1=bp_2009[bp_2009.smoke==1]
bp_2009_2=bp_2009[bp_2009.smoke==2]
bp_2009_3=bp_2009[bp_2009.smoke==3]
bp_2009_mean_1 = bp_2009_1.groupby('age').mean()
bp_2009_mean_2 = bp_2009_2.groupby('age').mean()
bp_2009_mean_3 = bp_2009_3.groupby('age').mean()
bp_2009_mean_1 = bp_2009_mean_1.rename(columns={'b_press_contract': 'No_smoke_contract', 'b_press_relax': 'No_smoke_relax'})
bp_2009_mean_2 = bp_2009_mean_2.rename(columns={'b_press_contract': 'Server_smoke_contract', 'b_press_relax': 'Server_smoke_relax'})
bp_2009_mean_3 = bp_2009_mean_3.rename(columns={'b_press_contract': 'smoke_contract', 'b_press_relax': 'smoke_relax'})
bp_2009_mean_1 = bp_2009_mean_1.drop('smoke', axis = 1)
bp_2009_mean_2 = bp_2009_mean_2.drop('smoke', axis = 1)
bp_2009_mean_3 = bp_2009_mean_3.drop('smoke', axis = 1)
c_bp_mean_2009_smoke = pd.concat([bp_2009_mean_1, bp_2009_mean_2, bp_2009_mean_3], axis=1, join='inner')
c_bp_mean_2009_smoke.plot(linestyle = '-', marker = 'o', figsize = (10, 5))
```
