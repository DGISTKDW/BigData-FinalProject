```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
health2013 = pd.read_csv('NHIS_OPEN_GJ_2013.csv')
health2013.head()
```

```python
health2014 = pd.read_csv('NHIS_OPEN_GJ_2014.csv')
health2014.head()
```

```python
health2015 = pd.read_csv('NHIS_OPEN_GJ_2015.csv')
health2015.head()
```

```python
new=pd.concat([health2013,health2014,health2015])
```

```python
sexbs13=health2013[['sex','b_sugar']]
sexbsugar13=sexbs13.groupby('sex').mean()
sexbsugar13
```

```python
sexage13=health2013[['sex','age','b_sugar']]
sexage13_w=sexage13[sexage13['sex']==2]
sexage13_m=sexage13[sexage13['sex']==1]
sexage13_w=sexage13_w[sexage13_w['age']==11]
sexage13_m=sexage13_m[sexage13_m['age']==11]
sexage13_1=sexage13_w.groupby('age').mean()
sexage13_2=sexage13_m.groupby('age').mean()
result=pd.merge(sexage13_1,sexage13_2, left_index=True, right_index=True)
result
```

```python
sexage13=health2013[['sex','age','height','b_sugar']]
sexage13_w=sexage13[sexage13['sex']==2]
sexage13_m=sexage13[sexage13['sex']==1]
sexage13_w=sexage13_w[sexage13_w['age']==11]
sexage13_m=sexage13_m[sexage13_m['age']==11]
sexage13_w_155=sexage13_w[sexage13_w['height']==155]
sexage13_m_155=sexage13_m[sexage13_m['height']==155]
sexage13_m_155=sexage13_m_155.groupby('age').mean()
sexage13_w_155=sexage13_w_155.groupby('age').mean()
result1=pd.merge(sexage13_w_155,sexage13_m_155, left_index=True, right_index=True)
result1
```
